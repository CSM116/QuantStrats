```python
import yfinance as yf
import pandas as pd
from datetime import date

# List of Dow Jones (DJIA) constituents
tickers = [
    "AAPL", "AMGN", "AXP", "BA", "CAT", "CSCO", "CVX", "DIS", "DOW", "GS",
    "HD", "HON", "IBM", "INTC", "JNJ", "JPM", "KO", "MCD", "MMM", "MRK",
    "MSFT", "NKE", "PG", "TRV", "UNH", "V", "VZ", "WBA", "WMT"
]

# Define the date range
from_date = date(1928, 1, 1) 
to_date = date.today()

# Download historical data
data = yf.download(tickers, from_date, to_date, interval="1d", auto_adjust=False)
data = data.swaplevel(axis=1)
# Save to CSV
# data.to_csv("DJIA_Historical_Data.csv")

# Display the first few rows
display(data.head())
```
